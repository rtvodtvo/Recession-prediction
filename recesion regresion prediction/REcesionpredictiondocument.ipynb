## 1. Motivácia
<p>
        Stále počúvame, že sa blíži recesia. Technická definícia recesie je medziročný pokles reálneho štvrťročného hrubého domáceho produktu (HDP) v aspoň dvoch 
    po sebe nasledujúcich štvrťrokoch. Pre obyčajného človeka recesia znamená: riziko straty práce, málo voľných pracovných pozícií, prepad hodnoty jeho investícií,
    drahé požičavanie peňazí, drahšie existujúce dlhy a nízka kúpyschopnosť. Recesia v USA do veľkej mieri ovplyvňuje aj nás Slovákov. Slovensko a celá Európa je závislá 
    na importe. USA je najväčšia ekonomika na svete. Posledná recesia v USA vyvolala kolaps globálneho obchodu, ktorý sa prepadol o 15% medzi rokmi 2008 a 2009.<br/>
    Osobne investujem do indexových fondov a každý Slovák, ktorý si zvolil sporenie do 3. piliera investuje do nich tiež. Americký akciový trh sa prepadne v recesii
    v priemere o 32%. Predčasná predikcia poklesu HDP nám  dáva možnosť pripraviť sa na recesiu. Medzi všeobecné kroky prípravy sa pokladá: splatenie existujúceho dlhu, 
    odkladanie peňazí do núdzového fondu, diverzifikovanie príjmov a ukázanie zamestnávateľovi svoju nenahraditeľnosť v práci.<br/><br/><br/><br/><br/><br/><br/>
</p><br/><br/><br/>

## 2. Metodológia
 <p> 
    Predikcia HDP predstavuje typ analýzy, ktorá spadá do hĺbkovej analýzy dát. Hĺbková analýza dát alebo Data mining je proces objavovania vzorcov, ktoré vedú k 
    použiteľným znalostiam z veľkých súborov dát prostredníctvom jednej, alebo viacerých tradičných techník dolovania dát.
    Metodológia s ktorou mám najviac skúseností je CRISP-DM(The CRoss Industry Standard Process for Data Mining). Je to najpoužívanejší analytický 
    model a skladá sa zo šiestich fáz:
    </p>
    
* <b>1. Pochopenie cieľa </b>: Aké sú ciele a požiadavky projektu ?
* <b>2. Pochopenie dát  </b>:Aké údaje máme/potrebujeme? Sú tieto dáta čisté?
* <b>3. Príprava dát </b>: Ako predpracujeme údaje pre modelovanie?
* <b>4. Modelovanie </b>: Aké techniky modelovania by sme mali použiť?
* <b>5. Vyhodnotenie výsledkov </b>: Ktorý model najlepšie spĺňa zadané ciele?
* <b>6. Nasadenie </b>: Ako majú zainteresované strany prístup k výsledkom?


## 3. Pochopenie cieľa

#### 3.1 Business KPI(Kľúčové ukazovateľe výkonnosti)
* Vzhľadom na fakt, že recesiu chápeme ako pokles štvrťročného hrubého domáceho produktu (HDP) v aspoň dvoch po sebe nasledujúcich štvrťrokoch, cieľ
tohto projektu je predikovanie HDP na nasledujúcich 6 mesiacov od 5.1.2023 do 11.1.2023.

#### 3.1 Ciele Data miningu

* Zostaviť prediktívny model s koeficientom determinácie(„R kvadrát“) väčším ako 0.95
    <p> 
    Koeficient determinácie môže nadobúdať hodnoty maximálne 1, čo znamená dokonalú predikciu hodnôt závislej premennej. Naopak hodnota 0 znamená,
že model neprináša pre poznanie závislej premennej žiadnu informáciu, je úplne neužitočný.
   </p> 
* Zostaviť prediktívny model HDP v USA s strednou kvadratickou odchýlkov(RMSE) menšou ako 300. 
    <p> 
    (RMSE) je často používaná miera rozdielov medzi reálnymi pozorovanými hodnotami a hodnotami predpovedanými modelom.
    </p>

## 5. Príprava dát

Príprava dát zahŕňa riešenie hlavných nedostatkov dát z časti 4.3 Overenie kvality Dát. Dáta budú kvalitné ak:
   * Všetky atribúty budú opisovať jedno časové obdobie(väčšina atribútov opisovala stav za 1 mesiac preto som vybral mesiac)
   * Všetky atribúty budú mať rovnaké množstvo záznamov(dáta nebudú mať chýbajúce hodnoty)
   * Všetky atribúty budú aktuálne, posledný záznam bude mať dátum aspoň 1.4.2023
   * Všetky atribúty budú v správnom formáte pre aplikáciu metód strojového učenia

#### 5.2 Čistenie dát
##### Agregácia 
 Atribúty bank_credit, Continued Claims, M1 a M2 mali podobu týždenných záznamov. Pre integrovanie do tabuľky s mesačnými záznamami ich bolo potrebné
 transformovať na mesačné dáta. Pomocou agregačnej funkcie som spriemeroval týždenné dáta za daný mesiac a priradil som ich k danemu mesiaci.<br/>
 

##### Interpolácia chýbajúcich hodnôt
Dáta opisujúce akciový index obsahujú záznamy z každého pracovného dňa. Pri akciových trhoch použitie aggregačnej funkcie nedáva zmysel, lebo trhy vedia byť pomerne volatilné, a to by skreslilo priemernú hodnotu. Lepší nápad je použiť otváraciu hodnotu indexu za každý mesiac(hodnotu indexu v prvý deň mesiaca). Atribúty opisujúce akciový trh majú chýbajúce hodnoty, kvôli tomu, že burza je zatvorená počas sviatkov a cez víkend. Preto je potrebné doplniť hodnoty za 1. deň mesiaca kedy hodnota kvôli sviatku alebo víkendu chýba.
Atribút GDP mal podobu kvartálnych záznamov. V tabuľke s mesačnými záznamami by teda mal tento atribút 75% prázdnych hodnôt. Doplnenie chýbajúcich hodnôt a týchto prípadoch by sa dalo vyriešiť 2 spôsobmi. Doplnenie chýbajúcej hodnoty nasledujúcou hodnotou alebo doplnenie pomocou štatistickej metódy
Ja som doplnil hodnoty na základe interpolácie. Interpolácia je štatistická metóda, pomocou ktorej sa súvisiace známe hodnoty používajú na odhad neznámej. k existuje všeobecný  trend v množine údajových bodov, je možné primerane odhadnúť hodnotu chýbajúcich hodnôt.<br/>

## 6. Modelovanie
##### Predspracovanie dát
Prvý krok pri modelovaní je vytvorenie tréningových a testovacích dát. Tréningové údaje a testovacie údaje pochádzajú z rovnakého pôvodného súboru údajov. Pôvodné záznamy sú náhodne rozdelené v určenom pomere na 2 nezávislé množiny. Pri dostatku dát je najpoužívanejší pomer 70% trénovacie dáta / 30% testovacie dáta. Na trénovacích dátach sa optimalizujú parametre modelov a na testovacích dátach sa meria presnosť modelov. 
##### Výber modelov
Lineárna regresia je metóda strojového učenia, ktorá nájde lineárnu rovnicu, ktorá najlepšie popisuje koreláciu nezávislých premenných so závislou premennou. Nezávislé   <br/>
premenné sú finančné indikátori a závislá premenná je hodnota HDP v nasledujúcich mesiacoch.<br/>
Pre modelovanie som zvolil 3 modely(Lineárna regresia, Bayesovská lineárna regresia, Lasso regresia), ktoré rôznym spôsobom hľadajú lineárne rovnice medzi premennými.

## 7. Záver

Všetky modely preukázali dostatočné výsledky, pretože splnili ciele dataminingu z kapitoly 3.1. Na dosiahnutie business cieľa bolo potrebné natrénovať 18 modelov(3 typy modelov pre každý mesiac od 5.1.2023 do 11.1.2023). Lineárna, Lasso a Bayesovská regresia dobre reprezentujú vzťah medzi finančnými indikátormi a HDP, pretože všetky modely dosahovali koeficient determinácie väčší ako 0.999. Všetky modely strácali presnosť v dlhodobom horizonte. 2/3 modelov predikovali stagnáciu alebo pokles HDP. Index spotrebiteľských cien, HDP aktuálneho mesiaca a Sadzba federálnych fondov mali najväčšiu váhu pri predikcii. Tieto 3 indikátory sú zhodou okolností najsledovanejšie makroekonomické indikátory na akciovom trhu. Z pohľadu ekonomiky sú aktuálne hodnoty týchto indikátorov negatívne. Inflácia je stála vysoká, pohybuje sa na 5% úrovni. HDP má negatívny trend, z 2% zo začiatku roka je aktuálne na 1% úrovni. Sadzba federálnych fondov sa blíži ku vrcholu tohto ekonomického cyklu. Predikcia modelov sa zhoduje s predikciou Federálneho rezervného systému, ktorý predpokladá, že USA ekonomika upadne do konca roka do „miernej recesie.
